**2.5** Consider a two (equiprobable) class, one-dimensional problem with samples distributed according to the Rayleigh pdf in each class, that is,
$$
p(x|\omega_i) =
\begin{cases}
\frac{x}{\sigma_i^2} \exp\left(\frac{-x^2}{2\sigma_i^2}\right) & x \ge 0 \\
0 & x < 0
\end{cases}
$$
Compute the decision boundary point $g(x) = 0$.

We need to solve $p(x|\omega_i) = p(x|\omega_j)$.
$$\frac{x}{\sigma_i^2} \exp\left(\frac{-x^2}{2\sigma_i^2}\right) = \frac{x}{\sigma_j^2} \exp\left(\frac{-x^2}{2\sigma_j^2}\right)$$
$$\frac{\sigma_j^2}{\sigma_i^2} = \exp\left(\frac{-x^2}{2\sigma_i^2} + \frac{x^2}{2\sigma_j^2}\right)$$
$$\ln\left(\frac{\sigma_j^2}{\sigma_i^2}\right) = \frac{-x^2}{2\sigma_i^2} + \frac{x^2}{2\sigma_j^2}$$
$$\ln\left(\frac{\sigma_j^2}{\sigma_i^2}\right) = \frac{-\sigma^2_jx^2 + \sigma^2_ix^2}{2\sigma_i^2\sigma^2_j}$$
$$\frac{2\sigma_i^2\sigma^2_j\ln\left(\frac{\sigma_j^2}{\sigma_i^2}\right)}{\sigma^2_i - \sigma^2_j} = x^2$$
$$\left(\frac{2\sigma_i^2\sigma^2_j\ln\left(\frac{\sigma_j^2}{\sigma_i^2}\right)}{\sigma^2_i - \sigma^2_j}\right)^{1/2} = x$$


**2.7** In a three-class, two-dimensional problem the feature vectors in each class are normally distributed with covariance matrix
$$
\Sigma =
\begin{bmatrix}
1.2 & 0.4 \\
0.4 & 1.8
\end{bmatrix}
$$
The mean vectors for each class are $[0.1, 0.1]^T$, $[2.1, 1.9]^T$, $[-1.5, 2.0]^T$.
Assuming that the classes are equiprobable, (a) classify the feature vector $[1.6, 1.5]^T$ according to the Bayes minimum error probability classifier; (b) draw the curves of equal Mahalanobis distance from $[2.1, 1.9]^T$.


(a) Using the discriminant function, with all classes being equiprobable,
$$g_1(1.6, 1.5) = \frac{1}{2} \left((1.6,1.5)-(0.1,0.1)\right)^T \begin{bmatrix} 0.9 & -0.2 \\ -0.2 & 0.6 \end{bmatrix} \left((1.6,1.5)-(0.1,0.1)\right) = 1.1805$$
$$g_2(1.6,1.5) = \frac{1}{2} \left((1.6,1.5)-(2.1,1.9)\right)^T \begin{bmatrix} 0.9 & -0.2 \\ -0.2 & 0.6 \end{bmatrix} \left((1.6,1.5)-(2.1,1.9)\right)=0.1205$$
$$g_3(1.6,1.5) = \frac{1}{2} \left((1.6,1.5)-(-1.5,2.0)\right)^T \begin{bmatrix} 0.9 & -0.2 \\ -0.2 & 0.6 \end{bmatrix} \left((1.6,1.5)-(-1.5,2.0)\right)=4.7095$$
Since the discriminant is smallest for Class 2, we conclude that the feature vector (1.6,1.5) is part of Class 2.
